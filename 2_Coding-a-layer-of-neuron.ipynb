```python
import numpy as np
```

```python
X = [[1,2,3]] #input
W = [[0.1, 0.3, 0.2],
     [0.4, 0.2, 0.1],
     [0.01, -0.3, -0.6]] #Weights
b = [0.01, -0.03, 0.15] #bias

#Converting to matrix
X = np.array((X))
W = np.array((W)) 

Y = np.dot(X, W.T) + b

print(Y)
```
