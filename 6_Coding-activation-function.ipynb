# ReLU Function

```python
import numpy as np
```

```python
# Suppose from a particular layer the output is O

def ReLU(inputs):
    output = np.maximum(0, inputs)
    return output

O = [[0.5,-0.4, 0.3],
     [0.45,0.4, -0.3],
     [-0.35, 0.2, 0.15]]

O = np.array((O))

final_output = ReLU(O)

print(final_output)
    
```

# Softmax Function

```python
def Softmax(inputs):
    #unnormalized probabilities
    value = inputs - np.max(inputs, axis = 1, keepdims = True)
    exp_val = np.exp(value)
    #Generating Probabilities with softmax formula
    Output = exp_val / np.sum(exp_val, axis = 1, keepdims = True)
    return Output

O = [[0.5,-0.4, 0.3],
     [0.45,0.4, -0.3],
     [-0.35, 0.2, 0.15]]

O = np.array((O))

final_output = Softmax(O)

print(final_output)
```
