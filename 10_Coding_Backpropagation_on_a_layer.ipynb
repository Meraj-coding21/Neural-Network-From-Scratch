```python
import numpy as np
```

```python
class Dense_Layer:
    def __init__(self, n_neurons, n_inputs):
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    def forward_pass(self, inputs):
        self.inputs = inputs
        self.output = np.dot(inputs, self.weights) + self.biases
        return self.output
    def backward_pass(self, dl_dz, learning_rate):
        self.dweights = np.dot(self.inputs.T, dl_dz)
        self.dbiases = np.sum(dl_dz, axis = 0, keepdims = True)

        self.weights = self.weights - learning_rate * self.dweights
        self.biases = self.biases - learning_rate * self.dbiases
```

```python
class ReLU:
    def forward_pass(self, inputs):
        self.output = np.maximum(0, inputs)
        return self.output
    def backward_pass(self, dl_da):
        da_dz = np.where(self.output > 0 , 1, 0)
        return da_dz * dl_da 
```

```python
X = [[1, 2, 3, 4]]
X = np.array(X)

#forward_pass

Layer1 = Dense_Layer(3,4)
activation = ReLU()

z = Layer1.forward_pass(X)
a = activation.forward_pass(z)

total_loss = np.sum(a)

print(f"total loss : {total_loss}")

#backward pass

dl_da = np.ones_like(a)

dl_dz = activation.backward_pass(dl_da)

Layer1.backward_pass(dl_dz, 0.01)

print("One forward and backward pass complete")
```
