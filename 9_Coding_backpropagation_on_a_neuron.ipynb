```python
import numpy as np
```

```python
X = [1,2,3]
W = [2,2,3]
B = [1]
target = 0  #our target is to make loss to be 0

X = np.array((X)) #making it 2D, in this single neuron problem 1D will work anyways but in real scenario we will always be working in batches and multiple inputs so we need numpy array then. For better practice did this
W = np.array((W))
B = np.array((B))

def forward_pass(X,W,B):
    summ = np.dot(X, W.T) + B
    relu_sum = np.maximum(0, summ)
    loss = (relu_sum - target) ** 2
    return summ, relu_sum, loss

def backpropagation(X, W, B, summ, relu_sum, loss):
    learning_rate = 0.001
    
    d_loss_relu_sum = 2 * relu_sum  # our target is zero. if there was a target value then it would have been 2 * (relu_sum - target)
    
    if relu_sum > 0 :               
        d_relu_sum_summ = 1
    else:
        d_relu_sum_summ = 0

    first_two_terms = d_loss_relu_sum * d_relu_sum_summ

    dw = first_two_terms * X
    db = first_two_terms * 1

    #update weights and bias
    W = W - learning_rate * dw
    B = B - learning_rate * db

    return W,B
    

summ, relu_sum, loss = forward_pass(X,W,B)
print(loss)

for i in range(200):
    W, B = backpropagation(X, W, B, summ, relu_sum, loss)
    summ, relu_sum, loss = forward_pass(X,W,B)
    if i == 0:
        print(loss) # 240.87, exactly what we calculated in notes
print(loss) 
```
