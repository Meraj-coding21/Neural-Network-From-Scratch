```python
import numpy as np
```

```python
X = [[1,2,3],
    [5,6,7]] #input
W1 = [[0.1, 0.3, 0.2],
     [0.4, 0.2, 0.1],
     [0.01, -0.3, -0.6]] #Weights1

W2 = [[0.1, 0.2, -0.2],
     [0.4, -0.4, 0.1],
     [0.11, -0.3, -0.5]] #Weights2

b1 = [0.01, -0.03, 0.15] #bias1
b2 = [0.02, -0.04, -0.32]  #bias2

#Converting to matrix
X = np.array((X))
W1 = np.array((W1)) 
W2 = np.array((W2))

O = np.dot(X, W1.T) + b1
Y = np.dot(O, W2.T) + b2

print(Y)
```
