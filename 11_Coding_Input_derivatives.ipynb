```python
import numpy as np
```

```python
class Dense_Layer:
    def __init__(self, n_neurons, n_inputs):
        self.weights = 0.01 * np.random.randn(n_inputs, n_neurons)
        self.biases = np.zeros((1, n_neurons))
    def forward_pass(self, inputs):
        self.inputs = inputs
        self.outputs = np.dot(inputs, self.weights) + self.biases
        return self.outputs
    def backward_pass(self, dl_dz, learning_rate):
        self.dweights = np.dot(self.inputs.T,dl_dz)
        self.dbiases = np.sum(dl_dz, axis = 0, keepdims = True) # this is for batch of inputs
        self.dinputs = np.dot(dl_dz, self.weights.T) # derivatives for input

        self.weights = self.weights - learning_rate * self.dweights
        self.biases = self.biases - learning_rate * self.dbiases

        return self.dinputs
```

```python
class ReLU:
    def forward_pass(self, inputs):
        self.outputs = np.maximum(inputs, 0)
        return self.outputs
    def backward_pass(self, dl_da):
        da_dz = np.where(self.outputs > 0, 1, 0)
        return dl_da * da_dz           
```

```python
X = [[1,2,3,4]]
X = np.array(X)
print(X)

#forward pass

Layer1 = Dense_Layer(3,X.shape[1])
activation1 = ReLU()

z = Layer1.forward_pass(X)
a = activation1.forward_pass(z)

y_pred = np.sum(a)
loss = y_pred ** 2

print(f"loss:{loss}")

#backward pass

dl_dy = 2 * y_pred
dl_da = np.ones_like(a)
dl_da = dl_da * dl_dy

dl_dz = activation1.backward_pass(dl_da)

updated_inputs = Layer1.backward_pass(dl_dz, 0.01)

print(updated_inputs)
```
